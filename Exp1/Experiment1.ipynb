```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/download.webp')
```

```python
if img is None:
    print("Image not found. Check file name/path.")
else:
    # Convert BGR → RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display
    plt.imshow(img_rgb)
    plt.title("Acquired Image")
    plt.axis('off')
    plt.show()
```
