```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/tree.jpg',0)
```

```python
gray=np.float32(img)
```

```python
corner=cv2.cornerHarris(gray,2,3,0.04)
img_float=np.float32(img)
corners=cv2.cornerHarris(img_float,2,3,0.04)
```

```python
cv2.imwrite("corner.jpg",corner)
cv2.imwrite("corners.jpg",corners)
```

```python
Pimg=cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
Pimg[corners>0.01*corners.max()]=[0,0,255]
```

```python
cv2.imwrite("Pimg.jpg",Pimg)
```
