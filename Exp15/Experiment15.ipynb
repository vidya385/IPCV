```python
import cv2
```

```python
cap = cv2.VideoCapture('/content/motion.mp4')

if not cap.isOpened():
    raise ValueError("Video not found. Upload video first")
```

```python
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) or 640
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) or 480
fps    = cap.get(cv2.CAP_PROP_FPS)
fps    = fps if fps > 0 else 20
```

```python
out = cv2.VideoWriter('/content/output_bw_tracking.mp4',
                      cv2.VideoWriter_fourcc(*'mp4v'),
                      fps, (width, height))
```

```python
ret, frame1 = cap.read()
ret2, frame2 = cap.read()
```

```python
if not ret or not ret2:
    raise ValueError(" Not enough frames")
```

```python
while ret2:
    # Frame difference
    diff = cv2.absdiff(frame1, frame2)

    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)

    # Binary mask (WHITE object, BLACK background)
    _, thresh = cv2.threshold(blur, 25, 255, cv2.THRESH_BINARY)

    # Convert to 3-channel
    mask_colored = cv2.cvtColor(thresh, cv2.COLOR_GRAY2BGR)

    # Write frame
    out.write(mask_colored)

    frame1 = frame2
    ret2, frame2 = cap.read()
```

```python
cap.release()
out.release()
```
